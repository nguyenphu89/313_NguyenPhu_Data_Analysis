```python
import numpy as np

# Tạo mảng toàn số 0 (2x2)
a = np.zeros((2, 2))
print(a)
# [[0. 0.]
#  [0. 0.]]

# Tạo mảng toàn số 1 (1x2)
b = np.ones((1, 2))
print(b)
# [[1. 1.]]

# Tạo mảng hằng số (2x2), giá trị = 7
c = np.full((2, 2), 7)
print(c)
# [[7. 7.]
#  [7. 7.]]

# Tạo ma trận đơn vị 2x2
d = np.eye(2)
print(d)
# [[1. 0.]
#  [0. 1.]]

# Tạo mảng 2x2 với giá trị ngẫu nhiên [0,1)
e = np.random.random((2, 2))
print(e)
# [[0.91940167 0.08143941]
#  [0.68744134 0.87236687]]
```

###Python Finance Application with Pandas

```python
import pandas as pd

# Tạo DataFrame để lưu trữ giao dịch
columns = ['Date', 'Description', 'Amount']
ledger = pd.DataFrame(columns=columns)

# Hàm thêm giao dịch
def add_transaction(date, description, amount):
    global ledger
    new_transaction = pd.DataFrame(
        [[date, description, amount]],
        columns=columns
    )
    ledger = pd.concat([ledger, new_transaction], ignore_index=True)

# Hàm hiển thị toàn bộ sổ giao dịch
def view_ledger():
    print(ledger)

# Hàm lấy số dư hiện tại
def get_balance():
    return ledger['Amount'].sum()

# Thêm một vài giao dịch mẫu
add_transaction('2023-11-01', 'Income', 1000)
add_transaction('2023-11-02', 'Groceries', -200)
add_transaction('2023-11-03', 'Utilities', -100)

# Xem sổ giao dịch
view_ledger()

# Kiểm tra số dư hiện tại
print("Current Balance:", get_balance())
```
